```python
import pandas as pd
import numpy as np
```

```python
train= pd.read_csv("dataset/Train.csv")
test= pd.read_csv("dataset/Test.csv")
sub= pd.read_csv("dataset/Sample Submission.csv")
```

```python
train.head()
```

```python
train.info()
```

```python
target= train["UnitPrice"]
```

```python
train.drop(["UnitPrice"],axis=1,inplace=True)
```

```python
all_data= pd.concat([train, test], ignore_index=True, axis=0)
```

```python
all_data['InvoiceDate'] = pd.to_datetime(all_data['InvoiceDate'])
```

```python
#creating features from datetime feature.
#all_data['month'] = all_data["InvoiceDate"].dt.month
#all_data['day'] = all_data["Datetime"].dt.day
all_data['year'] = all_data["InvoiceDate"].dt.year
all_data['quarter'] = all_data["InvoiceDate"].dt.quarter
#all_data["dayofyear"] = all_data["InvoiceDate"].dt.dayofyear
all_data["weekofyear"] = all_data["InvoiceDate"].dt.weekofyear


all_data['semester'] = np.where(all_data.quarter.isin([4,1]),1,2)
#all_data['is_weekend'] = np.where(all_data['dayofweek'].isin([5,6]),1,0)
```

```python
all_data.drop(["InvoiceDate"],axis=1,inplace=True)
```

```python
#test.drop(["InvoiceDate"],axis=1,inplace=True)
```

```python
all_data.head()
```

```python
#Let's do the correlation plot to check element to element co- relations
corr = all_data.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
dtrain= all_data[:train.shape[0]]
dtest= all_data[train.shape[0]:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_valid, y_train, y_valid= train_test_split(dtrain, target, test_size=0.33, random_state=40)
```

```python
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
from sklearn.tree import DecisionTreeRegressor
model= DecisionTreeRegressor()
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=200, min_samples_split=5,random_state=50 )
```

```python
model.fit(x_train, y_train)
```

```python
pred= model.predict(x_valid)
```

```python
from sklearn.metrics import mean_squared_error
rmse= mean_squared_error(y_valid, pred)**(1/2)
```

```python
y_pred= model.predict(dtest)
```

```python
df= pd.DataFrame()
df["UnitPrice"] = y_pred
```
